# Text cleaning
Remove html tags,  
remove special char,

# removing all html tags

```python
import re
clean= re.compile('<.*?>')
re.sub(clean,'',df.iloc[2].review)
```

```python
 def clean_tags(text):
        clean= re.compile('<.*?>')
        return re.sub(clean,'',text)
```

```python
df["review"]=df["review"].apply(clean_tags)
```

```python
df.head(2)
```

converting in lower case

# CONVERTING IN LOWER CASE

```python
def cnvrt_lower(text):
    return text.lower()
```

```python
df["review"]=df["review"].apply(cnvrt_lower)
```

removing special character

# REMOVING SPECIAL CHARACTER

```python
def remove_special(text):
    x=''
    for i in text:
        

        if  i.isalnum():
            x=x+i
        else:
            x=x+' '
    return x
```

```python
df["review"]=df["review"].apply(remove_special)
```

# REMOVING STOP WORDS

```python
import nltk 
```

```python
from nltk.corpus import stopwords
```

```python
def rmv_stopwords(text):
    x=[]
    for i in text.split():
        if i not in stopwords.words("english"):
            x.append(i)
    y=x[:]
    x.clear()
    return y
```

```python
df["review"]=df["review"].apply(rmv_stopwords)
```

# Stemming

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
#steming
y=[]
def stem_word(text):
    for i in text :
        y.append(ps.stem(i))
    z=y[:]
    y.clear()
    return z
```

```python
df["review"]=df["review"].apply(stem_word)
```

```python
def join_back(list_input):
    return " ".join(list_input)
```

```python
df["review"].apply(join_back)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
df["review"] = df["review"].apply(lambda x: ' '.join(x))
```
